## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("../data/company_sales_data.csv")
plt.figure(figsize=(8,5))
axis1=data["total_profit"]
axis0=data["month_number"]
plt.plot(axis0,axis1,scalex=True)
plt.xticks(range(1,13))
plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.title("Company profit per month")
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
plt.figure(figsize = (10,5))
Sales = data.iloc[:,1:7]
month_number = data['month_number']
Headers = list(Sales.columns.values)
for i in range(len(Headers)):
    plt.plot(month_number,Sales[Headers[i]],marker = 'o',label = Headers[i])
    plt.legend()
    plt.locator_params('x',nbins = 20)
    plt.xlabel('Month Number')
    plt.ylabel('Sale units')
plt.show()
```

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
import numpy as np
plt.figure(figsize=(10,5))
# set width of bars
barWidth = 0.25
 
# set heights of bars
bars1 = data['facecream']
bars2 = data['facewash']
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 

plt.bar(r1, bars1, color='blue', width=barWidth, edgecolor='white', label='Face Cream sales data')
plt.bar(r2, bars2, color='orange', width=barWidth, edgecolor='white', label='Face Wash sales data')

 

plt.xlabel('Month Number', fontweight='bold')
plt.ylabel('Sales units in number', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], data['month_number'])
plt.title("Facewash and Facecream sales data", fontweight='bold')

plt.legend()
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
faccream = data.loc[:,'facecream']
facwash = data.loc[:,'facewash']
tothpaste = data.loc[:,'toothpaste']
bathingsoap = data.loc[:,'bathingsoap']
shampoo  = data.loc[:,'shampoo']
moisturizer = data.loc[:,'moisturizer']
months = data.loc[:,'month_number']

plt.figure(figsize=(10,5))
plt.plot([],[],color='m', label='face Cream')
plt.plot([],[],color='c', label='Face wash')
plt.plot([],[],color='r', label='Tooth paste')
plt.plot([],[],color='k', label='Bathing soap')
plt.plot([],[],color='g', label='Shampoo')
plt.plot([],[],color='y', label='Moisturizer')

plt.stackplot(months, faccream, facwash, tothpaste, bathingsoap, shampoo, moisturizer, colors=['m','c','r','k','g','y'])
plt.legend(loc="upper left")
plt.xlabel('Month Number')
plt.ylabel('Sale units')
plt.title('All product sales data using stack plot', fontweight = 'bold')
plt.show()
```
